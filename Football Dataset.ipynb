### How would you create a new column to transform a column ‘Age’ from years to decades? (1 mark)

```python
df['Age_Decades'] = df['Age'].apply(lambda age: age/10)
```

```python
df
```

### Which Algeria player had the highest overall rating? (2 marks)

```python
df_algeria = df[df['Nationality']=='Algeria']
df_algeria.groupby('Name')['Overall'].max().sort_values(ascending = False).head(3)
```

### Which back had the highest rating for ‘Sliding tackle’? (2 marks)

```python
df_back = df[df['Preferred Positions Type']=='Back'].groupby('Name')['Sliding tackle'].max().sort_values(ascending= False)
df_back
```

### Which preferred position type of England has on average the highest overall rating? (2 marks)

```python
df[df['Nationality']=='England'].groupby('Preferred Positions Type')['Overall'].mean().sort_values(ascending=False)
```

### Brazil’s forwards have a higher average overall rating than the backs (1 mark)

```python
df[df['Nationality']=='Brazil'].groupby('Preferred Positions Type')['Overall'].mean().sort_values(ascending=False)
```

### Which country has the oldest player? (2 marks)

```python
df.groupby(['Nationality','Name'])['Age'].max().sort_values(ascending=False).head(3)
```

### Which of the following attributes is on average the lowest for goalkeepers? (3 marks)

```python
df.groupby('Preferred Positions Type')[['GK diving','GK kicking', 'GK handling', 'GK reflexes']].mean()
```

### Which preferred positions type has the most entries in this dataset? (3 marks)

```python
df.groupby('Preferred Positions Type')['Preferred Positions Type'].count().sort_values(ascending=False)
```

### Which player from Portugal, that is younger than 25, has the highest overall rating? (3 marks)

```python
df[(df['Nationality']=='Portugal') & (df['Age'] < 25)].groupby('Name')['Overall'].max().sort_values(ascending=False)
```
